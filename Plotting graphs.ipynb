PLOT OF THE LEFT BRIGHT RIGHT DARK

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('Measurements/Plot Crosssection left bright right dark 2.csv')

# Create plot
plt.figure(figsize=(10, 6), facecolor='white')  # Set figure size

# Plot data
for column in data.columns[1:]:  # Assuming the first column is x-axis data
    plt.plot(data[data.columns[0]], data[column], label=column)  # Assuming the first column is x-axis data

# Customize plot
plt.xlabel('Time [s]', fontsize = 14)
plt.ylabel('Average Intensity [%]', fontsize = 14)
plt.title('Average Intensity in the two channels', fontsize = 16)
plt.grid(True)

# Add a bit of margin to the x-axis limits
x_min = data[data.columns[0]].min() - 0.02 * (data[data.columns[0]].max() - data[data.columns[0]].min())
x_max = data[data.columns[0]].max() + 0.02 * (data[data.columns[0]].max() - data[data.columns[0]].min())
plt.xlim(x_min, x_max)

# Add a bit of margin to the y-axis limits
y_max = data.iloc[:, 1:].values.max() * 1.4  # Increase the space by 20%
plt.ylim(0, y_max)

# Add red vertical line at x=1.5
plt.axvline(x=1.5, color='red', linestyle='--')
# Annotate the line with "trigger"
plt.text(1.5, -0.05*y_max, 'Trigger', color='red', fontsize=12, ha='center')

# Add legend with Arial font
plt.legend(fontsize='medium', loc='upper right', fancybox=True, framealpha=1, shadow=True)
plt.gca().set_facecolor('white')  # Set background color

# Set font to Arial
plt.rcParams['font.family'] = 'Arial'

# Show plot
plt.show()
```
